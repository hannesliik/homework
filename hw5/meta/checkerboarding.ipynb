```python
import numpy as np
import matplotlib.pyplot as plt
def get_goal(is_evaluation=False, interval=5):
    leng = 10
    intervals = int(leng / interval )
    #print(intervals)
    x_bin = np.random.randint(0, intervals) * 2 - intervals 
    y_bin = np.random.randint(0, intervals) * 2 - intervals
    if is_evaluation:
        if np.random.randn() > 0:
            x_bin += 1
        else:
            y_bin += 1
    else:
        if np.random.randn() > 0:
            x_bin += 1
            y_bin += 1
    #x_bin /= intervals
    #y_bin /= intervals
    x = np.random.uniform(x_bin, (x_bin +  1) ) * (leng / intervals)
    y = np.random.uniform(y_bin, (y_bin +  1) ) * (leng / intervals)
    
    #x = x_bin
    #y = y_bin
    goal = np.array([x, y])
    return goal
    
def plot_goals(goals):
    plt.scatter(goals[:,0], goals[:,1], alpha=0.1)
    #plt.show()
```

```python
figure = plt.figure(dpi=200)
plot_goals(np.array([get_goal(False) for i in range(10000)]))
plot_goals(np.array([get_goal(True) for i in range(10000)]))
plt.show()
```
